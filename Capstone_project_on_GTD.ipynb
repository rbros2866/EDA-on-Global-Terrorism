This analysis can help to inform policy decisions, resource allocation, and counterterrorism strategies.

* Correlations between attack characteristics such as target type, attack type, and perpetrator group
* Temporal patterns of attacks over time, such as seasonal trends or changes in attack frequency over time
* Geospatial analysis of attacks using maps to visualize the distribution and concentration of attacks across different regions and countries.

### Variables Description 

**The variables in the Global Terrorism Database can be categorized into three types: event-level variables, group-level variables, and contextual variables.**

**Event-level variables:**

* **iyear**: Year of the incident.
* **imonth**: Month of the incident. 
* **iday**: Day of the incident.
* **country_txt**: Name of the country where the incident occurred.
* **region_txt**: Name of the region where the incident occurred.
* **city**: Name of the city where the incident occurred.
* **latitude**: Latitude of the city where the incident occurred.
* **longitude**: Longitude of the city where the incident occurred.
* **attacktype1_txt**: Type of attack.
* **nkill**: Number of people killed in the incident.
* **nwound**: Number of people wounded in the incident.
* **target1**: Primary target of the attack.
* **summary**: Summary of the incident.
* **gname**: Name of the group responsible for the attack.
* **targtype1_txt**: Type of target.
* **weaptype1_txt**: Type of weapon used in the attack.
* **motive**: Motive behind the attack.
* **eventid**: A unique identifier for the incident.

**Group-level variables:**

* **gname**: Name of the group responsible for the attack.
* **gsubname**: Subname of the group responsible for the attack.
* **gname2**: Secondary perpetrator group responsible for the attack.
* **gsubname2**: Subname of the secondary perpetrator group responsible for the attack.
* **gname3**: Tertiary perpetrator group responsible for the attack.
* **gsubname3**: Subname of the tertiary perpetrator group responsible for the attack.
* **guncertain1**: Indicates whether the group responsible for the attack is unknown or uncertain.

**Contextual variables:**

* **region_txt**: Name of the region where the incident occurred.
* **country_txt**: Name of the country where the incident occurred.
* **attacktype1_txt**: Type of attack.
* **targtype1_txt**: Type of target.
* **weaptype1_txt**: Type of weapon used in the attack.
* **success**: Indicates whether the attack was successful or not.
* **suicide**: Indicates whether the attack was a suicide attack.
* **crit1**: Indicates whether the incident meets criterion 1 (the incident must be intentional).
* **crit2**: Indicates whether the incident meets criterion 2 (the incident must have a political, economic, or social goal).
* **crit3**: Indicates whether the incident meets criterion 3 (the incident must be outside the context of legitimate warfare activities).
* **dbsource**: Source of the information on the incident.

## 3. ***Data Wrangling***

* The reason for selecting only few out of 135 columns for the Global Terrorism Database (GTD) dataset is to focus our analysis on the most relevant and informative variables. This allows us to streamline our exploratory data analysis (EDA) and gain insights that are more actionable and relevant to our research questions. Additionally, reducing the number of columns simplifies our data cleaning and processing efforts, making it easier to identify and handle missing or inconsistent values, and reducing the likelihood of overfitting in our statistical models. Furthermore, we can avoid redundancy and multicollinearity issues that can arise when working with a large number of highly correlated variables. Overall, selecting a subset of the most relevant variables from the GTD dataset is a common practice in EDA, helping us to effectively analyze the data and draw meaningful conclusions.*

**Dealing with missing values:**

 **1.**We can fill in missing values with a reasonable estimate based on the context of the data. For example, we can fill in null values in the **"Killed"**,**"Wounded"** and **"casualties"**columns with 0 since it's possible that no one was killed or wounded in some attacks. 

 **2.**The **"Target"** and **"city"** columns have a significant percentage of missing values, with 0.35% and 0.24%, respectively. Dropping these columns will not significantly impact the analysis and modeling of the data.

 **3.**For the **"Motive"** column, dropping the column is also an option since it has a high percentage of missing values (72.2%). Alternatively, we could impute the missing values by filling in the null values with a placeholder value such as "unknown".So use of paceholder would fetch good insights rather than dropping the rows.

**4.**Null values in the **"Summary"** column are replaced  with a placeholder value such as "No summary available". This can help to retain the information in other columns and avoid losing the entire row.

**5.**To impute missing **"latitude"** and **"longitude"** values,we can use the mean or median values. While this method can help us preserve the size of the dataset, it could also introduce bias and inaccuracies in the analysis if the imputed values are not representative of the actual values.Any way we are not going to work on lattitude and longitude in this EDA so we can keep those missing values as it is.

```python
# Fill null values in 'Killed' and 'Wounded' columns with 0
df1['Killed'].fillna(0, inplace=True)
df1['Wounded'].fillna(0, inplace=True)
df1['casualties'].fillna(0, inplace=True)
```

```python
# Drop the 'city' and 'target1' columns
df1.dropna(subset=['city', 'Target'], inplace=True)
```

```python
#fill null values with a placeholder value such as "unknown" in 'Motive' column
df1['Motive'].fillna('Unknown',inplace=True)
```

```python
#fill null values with a placeholder value such as "unknown" in 'Summary' column
df1['Summary'].fillna('No summary available', inplace=True)
```

```python
df1[df1[['latitude', 'longitude']].isnull().any(axis=1)][['city', 'Region', 'Country','latitude', 'longitude']]['city'].value_counts()
#from value counts it is learned that "Unknown" and "unknown" are considered different values.
```

```python
#from value counts it is learned that "Unknown" and "unknown" are considered different values.
#using only one value of 'Unknown' 
df1['city'] = df1['city'].apply(lambda x: 'Unknown' if x=='unknown' else x)
```

**Dealing with Data Types:**

**1.** **Year**, **Month**, **Day**: These columns have integer values and represent dates. We can convert them to datetime format using pd.to_datetime

**Note:** In the given dataset, there are several instances where the Year, Month, and Day columns are not consistent with each other, resulting in incorrect dates. For example, there are instances where the day value is greater than the number of days in the corresponding month, or the month value is greater than 12. Additionally, there are instances where the Year, Month, and Day columns are not populated correctly, resulting in invalid dates such as "19700000" or "19700001".Such cases handled by replacing incorrect dates with NaN

**2.** **Country**, **Region**, **City**, **AttackType**, **Target**, **Summary**, **Group**, **Target_type**, **Weapon_type**, **Motive:** These columns have categorical values and can be converted to categorical datatype using astype('category').

**3.** **Latitude**, **Longitude**, **Killed**, **Wounded**, **Casualties:** These columns have numerical values and can be kept as float datatype.

```python
print(df1.dtypes)
```

```python
#1
# replace incorrect dates with NaN
df1.loc[df1['Month'] == 0, 'Month'] = np.nan
df1.loc[df1['Day'] == 0, 'Day'] = np.nan

# Convert Year, Month, Day to datetime format
df1['Date'] = pd.to_datetime(df1[['Year', 'Month', 'Day']])
# Drop the Month, Day columns,we are not dropping "year" column as it contains valid data.It would be usefull for further analysis.
df1.drop(['Month', 'Day'], axis=1, inplace=True)
```

```python
#2
# Convert categorical columns to categorical datatype
df1['Country'] = df1['Country'].astype('category')
df1['Region'] = df1['Region'].astype('category')
df1['city'] = df1['city'].astype('category')
df1['AttackType'] = df1['AttackType'].astype('category')
df1['Target'] = df1['Target'].astype('category')
df1['Summary'] = df1['Summary'].astype('category')
df1['Group'] = df1['Group'].astype('category')
df1['Target_type'] = df1['Target_type'].astype('category')
df1['Weapon_type'] = df1['Weapon_type'].astype('category')
df1['Motive'] = df1['Motive'].astype('category')
```

# **Challenges & Future Work** 

The Global Terrorism Database (GTD) is a rich source of information for understanding the global patterns and trends in terrorist attacks. However, like any dataset, it has its own set of challenges that must be addressed during the exploratory data analysis (EDA) process. Some of these **challenges** include:

* Data Quality: The GTD is compiled from multiple sources, which can lead to inconsistencies and inaccuracies. As such, the data must be carefully cleaned and preprocessed before conducting any analysis.

* Missing Values: There may be missing values in the dataset that need to be imputed or handled in some other way to avoid bias in the analysis.

* Class Imbalance: The GTD is heavily skewed towards certain types of attacks and certain regions of the world. This can lead to bias in the analysis, and therefore it is important to address the issue of class imbalance during the EDA process.

* Time Series Analysis: The GTD is a time series dataset, and as such, time-related features need to be handled carefully during the EDA process. This includes trend analysis, seasonality analysis, and identifying anomalies.

* Dimensionality Reduction: The GTD is a large dataset with many features. It can be challenging to identify the most important features and reduce the dimensionality of the dataset to make it more manageable for analysis.

In terms of **future work,** there are many avenues for further exploration of the GTD dataset. Some potential areas of research include:

* Predictive Modeling: Using machine learning techniques to build predictive models that can identify the likelihood of future terrorist attacks.

* Spatial Analysis: Investigating the spatial patterns of terrorist attacks and identifying hotspots of activity.

* Network Analysis: Analyzing the social networks and organizational structures of terrorist groups to better understand their behavior and motivations.

* Temporal Analysis: Studying the temporal patterns of terrorist attacks and identifying temporal trends and cycles.

* Cross-Domain Analysis: Combining the GTD with other datasets, such as economic or social data, to identify correlations and causations between different factors and terrorism.
